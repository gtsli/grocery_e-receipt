```python
'''Takes cleaned branded data, splits each product title into a bucket based on the first letter of the product
So all titles beginning with A go into the A bucket, ect
'''
```

```python
import json
```

```python
'''
Use cleaned data that includes brands. Intended use case for this data is edit distance, which would work better with brands
'''
with open('../../synthetic_data/data/cleaned_branded_data.json') as f:
    data = json.load(f)
```

```python
buckets = [] #list containing each bucket, where buckets[0] = Dataset A, buckets[1] = Dataset B ...
for i in range(43):
    bu = []
    buckets.append(bu)
```

```python
#loop through all product titles, put into correct bucket (some unused buckets to make indexing nice)
for i in range(len(data)):
    letter = data[i][0].upper()
    ind = ord(letter) - 48
#     if (ind > 25 or (ind < 0 and not (ind > -50 and ind < -39))):
#         print(letter + " " + str(ind) + "  " + data[i])
    buckets[ind].append(data[i])
```

```python
#get rid of extra buckets
b = []
for i in range(10):
    b.append(buckets[i])
for i in range(17,43):
    b.append(buckets[i])
b
```

```python
'''
Save the buckets in a json file
'''
with open('bucket_data.json', 'w') as outfile:
    json.dump(b, outfile)
```
