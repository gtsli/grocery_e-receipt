# Data Cleaning

```python
DROP_COLS = [
    'NDB_Number',
    'data_source',
    'gtin_upc',
    'manufacturer',
    'date_modified',
    'date_available',
    'ingredients_english'
]
data = data.drop(DROP_COLS, axis=1)
```

```python
data.head()
```

```python
data['num_commas'] = data.apply(lambda x: len(x.long_name.split(",")), axis=1)
```

```python
len(data)
```

```python
data = data[data['num_commas'] < 4]
```

```python
len(data)
```

```python
data.head(25)
```

```python
def remove_brand_descriptor(x):
    x_ = x.split(",")
    
    if len(x_) > 1:
        if 'fl oz' in x_[1]:
            return x_[0].strip()
        else:
            return x_[1].strip()
    else:
        return x_[0].strip()
```

```python
data['cleaned_title'] = data.apply(lambda x: remove_brand_descriptor(x.long_name), axis=1)
```

```python
data
```

```python
new_titles = list(set(data.cleaned_title))
new_titles
```

```python
import json
with open('cleaned_data.json', 'w') as outfile:
    json.dump(new_titles, outfile)
```
